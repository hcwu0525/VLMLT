```python
from mhr.utils.utils import *
controlaug_file = "/mnt/petrelfs/songmingyang/songmingyang/data/llava_train/LLaVA-Instruct-150K/control/llavaft_control_llava13b_convs.jsonl"
llava_ft_file = "/mnt/petrelfs/songmingyang/songmingyang/data/llava_train/LLaVA-Instruct-150K/new/llava_v1_5_mix665k_newidx.json"

controlaug = process_jsonl(controlaug_file)
llavaft = load_json_file(llava_ft_file)

llavaft_dict = {str(item["id"]): item for item in llavaft}

```

```python
renewed_data = []
for item in controlaug:
    resume_id_name = item["id"].replace("_aug","").strip()
    if resume_id_name.startswith("VG_100K_2_"):
        resume_id_name = resume_id_name.replace("VG_100K_2_", "VG_100K_2/")
    elif resume_id_name.startswith("VG_100K_"):
        resume_id_name = resume_id_name.replace("VG_100K_", "VG_100K/")
        
    original_item = llavaft_dict.get(resume_id_name, None)
    if original_item:
        item["conversations"] = original_item["conversations"]
        renewed_data.append(item)
    else:
        print(item["id"])
```

```python
conversation_dict = {}
filtered_data = []
print(len(renewed_data))
for item in renewed_data:
    if conversation_dict.get(item["id"], None) is None:
        conversation_dict[item["id"]] = item
        filtered_data.append(item)
        
print(len(filtered_data))
    
```

```python
output_file = "/mnt/petrelfs/songmingyang/songmingyang/data/llava_train/LLaVA-Instruct-150K/control/control_originconv.jsonl"
write_jsonl(filtered_data,output_file)
```
