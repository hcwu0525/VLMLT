```python

from mhr.utils.utils import *
import os
origin_data_file="/mnt/petrelfs/songmingyang/songmingyang/data/llava_train/LLaVA-Instruct-150K/llava_v1_5_mix665k.json"
aug_image_dir = "/mnt/petrelfs/songmingyang/songmingyang/data/mm/imgs/llava_aug/aug_ft_200k"
```

```python
origin_data_data = load_json_file(origin_data_file)
aug_images = os.listdir(aug_image_dir)
image_relpath = "llava_aug/aug_ft_200k"
```

```python
origin_data_dict = {}
for item in origin_data_data:
    id = str(item["id"])
    if origin_data_dict.get(id) is None:
        origin_data_dict[id] = []
    origin_data_dict[id].append(item)

```

```python
aug_image_origin_cov=[]
for img in aug_images:
    img_id = img.split(".")[0]
    origin_image_id = str(img_id[:-4])
    origin_item = origin_data_dict.get(origin_image_id,None)
    if origin_item is None:
        if origin_image_id.startswith("VG_100K_2_"):
            origin_image_id = f"VG_100K_2/{origin_image_id[10:]}"
            origin_item = origin_data_dict.get(origin_image_id,None)
            if origin_item is None:
                print(f"origin_item is None for {img}, origin_image_id:{origin_image_id}")
                break
            elif len(origin_item) > 1:
                print(f"origin_item is None for {img}, origin_image_id:{origin_image_id}")
                break
        elif origin_image_id.startswith("VG_100K_"):
            origin_image_id = f"VG_100K/{origin_image_id[8:]}"
            origin_item = origin_data_dict.get(origin_image_id,None)
            if origin_item is None:
                print(f"origin_item is None for {img}, origin_image_id:{origin_image_id}")
                break
            elif len(origin_item) > 1:
                print(f"origin_item is None for {img}, origin_image_id:{origin_image_id}")
                break
        else:
            print(f"origin_item is None for {img}, origin_image_id:{origin_image_id}")
            break
    assert origin_item is not None and len(origin_item) == 1
    aug_image_origin_cov.append(dict(id=img_id,image=os.path.join(image_relpath,img),conversations=origin_item[0]["conversations"]))
    
        
```

```python
save_path = "/mnt/petrelfs/songmingyang/songmingyang/data/llava_train/LLaVA-Instruct-150K/reformed_data/dr_algo/aug/aug_img_with_origin_conv.json"
write_json_file(aug_image_origin_cov,save_path)
```

```python
from tqdm import tqdm
image_dir = "/mnt/petrelfs/songmingyang/songmingyang/data/mm/imgs"
for item in tqdm(aug_image_origin_cov):
    image_path = os.path.join(image_dir,item["image"])
    assert os.path.exists(image_path)
```
