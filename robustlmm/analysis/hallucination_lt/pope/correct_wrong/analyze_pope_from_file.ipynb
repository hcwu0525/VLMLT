```python
from mhr.utils.utils import process_jsonl
```

```python
pope_static_file = "/mnt/petrelfs/songmingyang/songmingyang/data/llava_train/benchmarks/analysis/pope/pope_object_rank.jsonl"
pope_data = process_jsonl(pope_static_file)
correct_min = []
correct_max = []
correct_mean = []
wrong_min = []
wrong_max = []
wrong_mean = []
for i in range(0, len(pope_data),6):
    correct = [x["correct"] for x in pope_data[i:i+6]]
    if sum(correct) < 6:
        wrong_max.append(pope_data[i]["max_rank"])
        wrong_min.append(pope_data[i]["min_rank"])
        wrong_mean.append(pope_data[i]["mean_rank"])
    else:
        correct_max.append(pope_data[i]["max_rank"])
        correct_min.append(pope_data[i]["min_rank"])
        correct_mean.append(pope_data[i]["mean_rank"])

print("correct min: ", sum(correct_min)/len(correct_min))
print("correct max: ", sum(correct_max)/len(correct_max))
print("correct mean: ", sum(correct_mean)/len(correct_mean))
print("wrong min: ", sum(wrong_min)/len(wrong_min))
print("wrong max: ", sum(wrong_max)/len(wrong_max))
print("wrong mean: ", sum(wrong_mean)/len(wrong_mean))
```
