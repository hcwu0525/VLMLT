```python
from mr_eval.utils.utils import *
```

```python
baseline_file = "/mnt/petrelfs/songmingyang/code/mm/robustLMM/robustlmm/eval_res/dr_algo/efficiency/test_flops/res/baseline_flops.jsonl"
aug_file = "/mnt/petrelfs/songmingyang/code/mm/robustLMM/robustlmm/eval_res/dr_algo/efficiency/test_flops/res/controlaug_flops.jsonl"

baseline_data = process_jsonl(baseline_file)
aug_data = process_jsonl(aug_file)
```

```python
def convert_to_human_readable_size(num):
    if num / 1e18 > 1:
        return f"{num / 1e18:.2f} E"
    elif num / 1e15 > 1:
        return f"{num / 1e15:.2f} P"
    elif num / 1e12 > 1:
        return f"{num / 1e12:.2f} T"
    elif num / 1e9 > 1:
        return f"{num / 1e9:.2f} B"
    elif num / 1e6 > 1:
        return f"{num / 1e6:.2f} M"
    elif num / 1e3 > 1:
        return f"{num / 1e3:.2f} K"
    else:
        return f"{num}"
```

```python
total = 0
for item in baseline_data:
    total += item["flops"]


print(f"Baseline total: {convert_to_human_readable_size(total)}")

total = 0
for item in aug_data:
    total += item["flops"]
    
print(f"Aug total: {convert_to_human_readable_size(total)}")
```

```python
len(baseline_data), len(aug_data)
```

```python
base_id_dict = {}
aug_id_dict = {}

renewed_baseline_data = []
renewed_aug_data = []

for item in baseline_data:
    if item["new_idx"] not in base_id_dict:
        base_id_dict[item["new_idx"]] = 1
        renewed_baseline_data.append(item)

for item in aug_data:
    if item["new_idx"] not in aug_id_dict:
        aug_id_dict[item["new_idx"]] = 1
        renewed_aug_data.append(item)
        
```

```python
len(renewed_baseline_data), len(renewed_aug_data)
```

```python
total = 0
for item in renewed_baseline_data:
    total += item["flops"]

total = total *3 *10
base_flops = convert_to_human_readable_size(total)


print(f"Baseline total: {base_flops}")

total = 0
for item in renewed_aug_data:
    total += item["flops"]
total = total *3 *10
print(f"Aug total: {convert_to_human_readable_size(total)}")
```
